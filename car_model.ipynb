```python
import pandas as pd
```

```python
df=pd.read_csv('Car Dataset Processed.csv')
```

```python
df.head()
```

```python
df['insurance_validity'].value_counts()
```

```python
d1={'Comprehensive':0,'Third Party insurance':1,'Zero Dep':2,'Third Party':1,'Not Available':3}
df['insurance_validity']=df['insurance_validity'].map(d1)
```

```python
df.head()
```

```python
df['fuel_type'].value_counts()
```

```python
d2={'Petrol':0,'Diesel':1,'CNG':2}
df['fuel_type']=df['fuel_type'].map(d2)
```

```python
df['fuel_type'].value_counts()
```

```python
df['ownsership'].value_counts()
```

```python
d3={'First Owner':1,'Second Owner':2,'Third Owner':3,'Fifth Owner':5,'Fourth Owner':4}
df['ownsership']=df['ownsership'].map(d3)
```

```python
df['ownsership'].value_counts()
```

```python
df['transmission'].value_counts()
```

```python
d4={'Manual':0,'Automatic':1}
df['transmission']=df['transmission'].map(d4)
```

```python
df.head()
```

```python
df['registration_year'] = df['registration_year'].str[-2:].astype(int)
```

```python
df.head()
```

```python
x=df[['insurance_validity','fuel_type','kms_driven','ownership','transmission']]
y=df['price(in lakhs)']
```

```python
x.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
models={'LinReg':LinearRegression(),
      'KNN-3':KNeighborsRegressor(),
      'SVM-Poly':SVR(kernel='poly')}
for name,model in models.items():
    model.fit(x,y)
    sc=model.score(x,y)
    print(name,sc)
```

```python
final_model=models['KNN-3']
```

```python
insurance_validity=input('insurance_validity')
fuel_type=input('fuel_type')
kms_driven=input('kms_driven')
ownsership=input('ownsership')
transmission=input('transmission')
insurance_validity=d1[insurance_validity]
fuel_type=d2[fuel_type]
transmission=d4[transmission]
ownsership=d3[ownsership]
kms_driven=int(kms_driven)



```

```python
test=[[insurance_validity,fuel_type,kms_driven,ownsership,transmission]]
final_model.predict(test)
```

```python
import pickle 
pickle.dump(final_model,open('final_model.pkl',"wb"))
```
