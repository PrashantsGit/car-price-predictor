```python
import pandas as pd
```

```python
df=pd.read_csv('Car Dataset Processed.csv')
```

```python
df.head()
```

```python
df['insurance_validity'].value_counts()
```

```python
df['insurance_validity']=df['insurance_validity'].map({'Comprehensive':0,'Third Party insurance':1,'Zero Dep':2,'Third Party':1,'Not Available':3})
```

```python
df.head()
```

```python
df['fuel_type'].value_counts()
```

```python
df['fuel_type']=df['fuel_type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
df['fuel_type'].value_counts()
```

```python
df['ownsership'].value_counts()
```

```python
df['ownsership']=df['ownsership'].map({'First Owner':0,'Second Owner':1,'Third Owner':2,'Fifth Owner':3})
```

```python
df['ownsership'].value_counts()
```

```python
df['transmission'].value_counts()
```

```python
df['transmission']=df['transmission'].map({'Manual':0,'Automatic':1})
```

```python
df.head()
```

```python
df['registration_year'] = df['registration_year'].str[-2:].astype(int)
```
